```python
h, w = 100, 150
img = np.zeros((h, w, 3), dtype=int)

img[:, 20:70, 2] = 255

pos_x = 80
pos_y = 10
rect_height = 30
rect_width = 60
img[pos_y:pos_y+rect_height, pos_x:pos_x+rect_width, :] = 255 # white Rectangle

plt.imshow(img)
plt.show()
```

```python
h, w = 100, 150
img = np.zeros((h, w, 3), dtype=int)

img[:h//2, w//2:, :] = 255  # First quarter (white)

img[:h//2, :w//2, 0] = 255  # Second quarter (Red)

img[h//2:, w//2:, 1] = 255  # Third quarter (Green)

img[h//2:, :w//2, 2] = 255  # Fourth quarter (Blue)

plt.imshow(img)
plt.show()
```

```python
img = cv2.imread('./images/baboon.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print(img.shape)
height, width, _ = img.shape

plt.imshow(img)
plt.show()
```

```python
plt.imshow(img[:height//2, 250:450])
plt.show()
```

```python
img[:height//2, width//2:, [0, 1]] = 0

plt.imshow(img)
plt.show()
```

```python
img[:height//2, width//2:, [0, 1]] = 0
img[:height//2, :width//2, [1, 2]] = 0
img[height//2:, width//2:, [0, 2]] = 0

plt.imshow(img)
plt.show()
```

```python
img[:height//2, width//2:, [0, 1]] = 0
img[:height//2, :width//2, [1, 2]] = 0
img[height//2:, width//2:, [0, 2]] = 0

# Calculating the average along the channel dimension (third dimension)
bw = np.mean(img[height//2:, :width//2, :], axis=2) # 2D Matrix, type of values are float
# Converting float values to int
bw = bw.astype(int) # 2D Matrix, type of values are int
# Add third dimension
bw = bw.reshape((height//2, width//2, 1)) # 3D Matrix

img[height//2:, :width//2, :] = bw

plt.imshow(img)
plt.show()
```

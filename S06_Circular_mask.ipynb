```python
height, width = 200, 300
img = np.zeros((height, width, 3), dtype=int)

X, Y = np.meshgrid(np.arange(width), np.arange(height))

print(X.shape, Y.shape)
print("X = \n", X)
print("Y = \n", Y)
```

```python
height, width = 200, 300
img = np.zeros((height, width, 3), dtype=int)

X, Y = np.meshgrid(np.arange(width), np.arange(height))

center_x = (width // 2)
center_y = (height // 2)

X -= center_x
Y -= center_y

print(X.shape, Y.shape)
print("X = \n", X)
print("Y = \n", Y)
```

```python
dist = np.sqrt(X**2 + Y**2)
```

```python
dist = np.sqrt(X**2 + Y**2)

mask1 = (dist <= 30)

plt.imshow(mask1, cmap='gray')
plt.show()
```

## Drawing circle

```python
dist = np.sqrt(X**2 + Y**2)

mask1 = (dist <= 30)

# Draw a Red circle with a radius of 30 pixels.
img[mask1, 0] = 255

plt.imshow(img)
plt.show()
```

## Drawing Ring

```python
# Ring Mask. R1 = 30 and R2 = 50
mask2 = (dist > 30) & (dist <= 50)

plt.imshow(mask2, cmap='gray')
plt.show()
```

```python
dist = np.sqrt(X**2 + Y**2)

mask1 = (dist <= 30)
# Draw a Red circle with a radius of 30 pixels.
img[mask1, 0] = 255

# Ring Mask. R1 = 30 and R2 = 50
mask2 = (dist > 30) & (dist <= 50)
# Draw a Green Ring with a width of 20 pixels.
img[mask2, 1] = 255

plt.imshow(img)
plt.show()
```

## Drawing square with a circular hole

```python
# Square mask (A=150 pixel)
A = 150
mask_square = (np.abs(X) <= A/2) & (np.abs(Y) <= A/2)

plt.imshow(mask_square, cmap='gray')
plt.show()
```

```python
A = 150
mask_square = (np.abs(X) <= A/2) & (np.abs(Y) <= A/2)
R = 50
mask_circle = (dist <= R)

# Square (A=150 pixel) with a circular hole (R = 50 pixel)
mask3 = mask_square & (~ mask_circle)

plt.imshow(mask3, cmap='gray')
plt.show()
```

```python
dist = np.sqrt(X**2 + Y**2)

mask1 = (dist <= 30)
# Draw a Red circle with a radius of 30 pixels.
img[mask1, 0] = 255

# Ring Mask. R1 = 20 and R2 = 50
mask2 = (dist > 30) & (dist <= 50)
# Draw a Green Ring with a width of 20 pixels.
img[mask2, 1] = 255

# Square (A=150 pixel) with a circular hole (R = 50 pixel)
A = 150
mask_square = (np.abs(X) <= A/2) & (np.abs(Y) <= A/2)
R = 50
mask_circle = (dist <= R)

mask3 = mask_square & (~ mask_circle)
img[mask3, 2] = 255

plt.imshow(img)
plt.show()
```

# Selecting Duration of Signal

```python
Fs = 1000 # Hz
t = np.arange(0., 1., 1/Fs)
f0 = 5
x = np.sin(2 * np.pi * f0 * t)

mask = (t >= 0.3) & (t <= 0.5)
x2 = x[mask]
t2 = t[mask]

plt.plot(t, x, label='x')
plt.plot(t2, x2, 'r--', label='x2')

plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()
```

# Thresholding

```python
Fs = 1000 # Hz
t = np.arange(0., 1., 1/Fs)
f0 = 5
x = np.sin(2 * np.pi * f0 * t)

threshold = 0.25
mask = (x < threshold)
x2 = x[mask]


plt.plot(t, x, label='x')
plt.axhline(y=threshold, c='r', label='threshold')
plt.plot(t, mask, 'k--', label='bloolean mask')
plt.xlabel('Time')
plt.grid(); plt.legend(); plt.show()
```

# Square wave Signal

```python
Fs = 1000 # Sampling Rate (Hz)
t = np.arange(0., 1., 1/Fs)
f0 = 5
x = np.sin(2 * np.pi * f0 * t)
s = (x > 0)
print(s.dtype)
s = s.astype(int)
print(s.dtype)
# s = (x > 0).astype(int)

plt.plot(t, x, label='sine wave')
plt.plot(t, s, 'k', label='square wave')
plt.title(f'Frequency = {f0} Hz'); plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()
```

```python
Fs = 1000 # Sampling Rate (Hz)
t = np.arange(0., 1., 1/Fs)
f0 = 3
T0 = 1 / f0     # period
cnt = (t % T0)
duty = 0.1
pwm = (cnt <= (duty * T0))

plt.plot(t, cnt, 'r')
plt.plot(t, pwm, 'k', label='PWM')
plt.title(f'Frequency = {f0} Hz, Duty = {duty * 100:0.2f}%')
plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()
```

# Saturation

```python
Fs = 1000 # Hz
t = np.arange(0., 1., 1/Fs)
f0 = 5
x = np.sin(2 * np.pi * f0 * t)

y = x.copy() # deep copy "x" to "y"

HIGH_THRESHOLD = 0.75
LOW_THRESHOLD = -0.5

mask_1 = (y > HIGH_THRESHOLD)
y[mask_1] = HIGH_THRESHOLD

mask_2 = (y < LOW_THRESHOLD)
y[mask_2] = LOW_THRESHOLD

plt.plot(t, x, label='x')
plt.plot(t, mask_1, 'k--', label='mask_1')
plt.plot(t, mask_2, 'r-.', label='mask_2')
plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()

# plt.plot(t, x, label='x')
# plt.plot(t, y, 'r--', label='y')
```

```python
plt.plot(t, x, label='x')
plt.plot(t, y, 'r--', label='y (saturated x)')
plt.title('Saturation'); plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()
```

```python
def saturate(x, high, low):
    y = x.copy() # deep copy "x" to "y"

    y[y >= high] = high
    y[y <= low] = low

    return y
```

```python
plt.plot(t, x, label='x')
plt.plot(t, saturate(x, .75, -0.5), 'r--', label='y (saturated x)')
plt.title('Saturation'); plt.xlabel('Time')
plt.legend(); plt.grid(); plt.show()
```
